```python
try:
    import cv2
    
    def draw_rectangle(event,x,y,flags,param):

        global pt1,pt2,in_process, border_drawen, clear

        
        if event == cv2.EVENT_LBUTTONDOWN:
            if in_process == False and border_drawen == False:
                pt1 = [x,y]
                pt2 = [x,y]
                in_process = True
            else:
                pt1 = [-1,-1]
                pt2 = [-1,-1]
                border_drawen = False
                in_process = False
                clear = True
        elif event == cv2.EVENT_MOUSEMOVE:
            if in_process == True:
                pt2 = [x,y]
        elif event == cv2.EVENT_LBUTTONUP:
            if clear == False:
                in_process = False
                border_drawen = True
            else:
                clear = False

    # Haven't drawn anything yet!

    pt1 = [-1,-1]
    pt2 = [-1,-1]
    in_process = False
    border_drawen = False
    clear = False
    cap = cv2.VideoCapture(0) 

    
    cv2.namedWindow('Frame')

    cv2.setMouseCallback('Frame', draw_rectangle) 


    while True:
        
        ret, frame = cap.read()
        
        if border_drawen == True and in_process == False:
            
            if pt2[0] < pt1[0]:
                pt2[0], pt1[0] = pt1[0], pt2[0]
            
            if pt2[1] < pt1[1]:
                pt2[1], pt1[1] = pt1[1], pt2[1]
            
            frame[pt1[1]: pt2[1],pt1[0]: pt2[0]] = cv2.blur(frame[pt1[1]: pt2[1],pt1[0]: pt2[0]],ksize=(50,50))
        
        cv2.imshow('Frame', frame)

        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    
    cap.release()
    cv2.destroyAllWindows()
    
except:
    cap.release()
    cv2.destroyAllWindows()
    print('Some problems appeared')
    
```
